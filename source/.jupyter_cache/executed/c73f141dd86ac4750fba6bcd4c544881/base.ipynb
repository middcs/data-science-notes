```python
#| fig-cap: '300 data points in the 2d plane, each of which has one of two labels.'
#| fig-cap-location: margin
#| label: fig-scatter

import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(precision = 3)
plt.style.use('seaborn-v0_8-whitegrid')
# from hidden.perceptron import perceptron_update, draw_line

def perceptron_data(n_points = 300, noise = 0.2):

    y = np.arange(n_points) >= int(n_points/2)
    X = y[:, None] + np.random.normal(0.0, noise, size = (n_points,2))
    X = np.append(X, np.ones((X.shape[0], 1)), 1)
    y = 2*y - 1

    return X, y

def plot_perceptron_data(X, y, ax):
    ax.scatter(X[:,0], X[:,1], s = 20,  c = y, facecolors = "none", edgecolors = "darkgrey", cmap = "BrBG", vmin = -2, vmax = 2, alpha = 0.5)
    ax.set(xlabel = r"$x_1$", ylabel = r"$x_2$")

fig, ax = plt.subplots(1, 1)
X, y = perceptron_data()
plot_perceptron_data(X, y, ax)
```

```python
#| code-fold: true
def draw_line(w, x_min, x_max, ax, **kwargs):
    w_ = w.flatten()
    x = np.linspace(x_min, x_max, 101)
    y = -(w_[0]*x + w_[2])/w_[1]
    l = ax.plot(x, y, **kwargs)
```

```python
#| fig-cap: Two candidate separating hyperplanes for our data. Which one would you choose?
#| fig-cap-location: margin
#| label: fig-scatter-with-separators

fig, ax = plt.subplots(1, 1, figsize = (4, 4))
plot_perceptron_data(X, y, ax)

w_0 = np.array([1, -1, 0])
w_1 = np.array([1,  1, -1]) 

draw_line(w_0, 0, 1, ax, color = "black", linestyle = "dashed", label = r"$w^{(0)}$")
draw_line(w_1, 0, 1, ax, color = "black", label = r"$w^{(1)}$")

l = ax.legend(ncol = 2)
```

```python
class Perceptron:

    def __init__(self):
        self.w = None

    def score(self, X):
        if self.w is None:
            self.w = np.random.rand(1, X.shape[1])

        return X@self.w.T

    def loss(self, X, y):
        s = self.score(X)
        
        return (1 - (s.flatten()*y > 0)).mean()

    def grad(self, X, y):
        s = self.score(X)
        # print(X*y, (s*y)<0)
        return -(X*y*(1.0*((s*y)<0))).mean(axis = 0)

class PerceptronOptimizer:

    def __init__(self, model):
        self.model = model 

    def step(self, X, y):
        n = X.shape[0]
        i = np.random.randint(n)
        x = X[[i],:]

        loss          = self.model.loss(x, y[i])
        self.model.w -= self.model.grad(x, y[i])

        return i, loss

p = Perceptron()
opt = PerceptronOptimizer(p)

p.loss(X, y)

# for j in range(100):
#     i, local_loss = opt.step(X, y)
#     # print(p.w)
#     print(p.loss(X, y))

# X[[2],]
```

```python
np.random.seed(1234)

p = Perceptron()
opt = PerceptronOptimizer(p)
p.loss(X, y)

plt.rcParams["figure.figsize"] = (8, 4)
fig, axarr = plt.subplots(2, 5, sharex = True, sharey = True)

current_ax = 0
loss = 1

while loss > 0:
    ax = axarr.ravel()[current_ax]

    old_w = p.w.copy()
    i, local_loss = opt.step(X, y)
    if local_loss > 0:
        plot_perceptron_data(X, y, ax)
        draw_line(old_w, x_min = -1, x_max = 2, ax = ax, color = "black", linestyle = "dashed")
        loss = p.loss(X, y)
        draw_line(p.w, x_min = -1, x_max = 2, ax = ax, color = "black")
        ax.scatter(X[i,0],X[i,1], color = "black", facecolors = "none", edgecolors = "black")
        # draw_line(w, -10, 10, ax, color = "black")
        ax.set_title(f"loss = {loss}")
        ax.set(xlim = (-1, 2), ylim = (-1, 2))
        current_ax += 1
plt.tight_layout()
```
